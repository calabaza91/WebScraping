```python
#Scraping from local files
from bs4 import BeautifulSoup as soup
```

```python
#Get filename
filename = 'Genome-Wikipedia.htm'
#Open file and store data in variable
f = open(filename)
#Return all data in readable method
#(Stored info, 'format.parser')
pageSoup = soup(f, 'html.parser')
print(pageSoup.h1)
```

```python
#Repeat same process in Web Scraping Notebook
genomeTable = pageSoup.findAll('table', {'class': 'wikitable sortable'})
```
